```python
# Merge CDC and ACS data file based on zip code
merged_df = pd.merge(df, filtered_df2, left_on='zcta5', right_on='zip_code', how='inner')
merged_df
```

```python
import numpy as np
from scipy.stats import linregress
import pandas as pd
import matplotlib.pyplot as plt

# Define age column renaming mapping
column_name_mapping = {
    'pop_18_19': 'Age 18-19',
    'pop_62_64': 'Age 62-64'
}

population_cols = list(column_name_mapping.keys())

# Define the lowest and highest age range columns
lowest_age_col = 'pop_18_19'
highest_age_col = 'pop_62_64'

# Extract the necessary data for the scatter plot
lowest_age_df = merged_df[merged_df[lowest_age_col].notnull()]
lowest_age_x = lowest_age_df[lowest_age_col]
lowest_age_y = lowest_age_df['casthma_crudeprev']

highest_age_df = merged_df[merged_df[highest_age_col].notnull()]
highest_age_x = highest_age_df[highest_age_col]
highest_age_y = highest_age_df['casthma_crudeprev']

# Create a scatter plot with two colors
plt.scatter(lowest_age_x, lowest_age_y, color='#00338D')
plt.scatter(highest_age_x, highest_age_y, color='#6D2077')

# Add labels and title to the plot
plt.xlabel(f'Number of People')
plt.ylabel('Average Asthma Prevalence')
plt.title(f'{column_name_mapping[lowest_age_col]} vs {column_name_mapping[highest_age_col]} with Average Asthma Prevalence')

# Add R-squared lines for each age range
slope, intercept, r_value, p_value, std_err = linregress(lowest_age_x, lowest_age_y)
r_squared = r_value**2
print(f'Low Age Range R-squared: {r_squared}')
plt.plot(lowest_age_x, slope * lowest_age_x + intercept, color='#00338D', label=f'{column_name_mapping[lowest_age_col]} (R-squared: {r_squared:.4f})')

slope, intercept, r_value, p_value, std_err = linregress(highest_age_x, highest_age_y)
r_squared = r_value**2
print(f'High Age Range R-squared: {r_squared}')
plt.plot(highest_age_x, slope * highest_age_x + intercept, color='#6D2077', label=f'{column_name_mapping[highest_age_col]} (R-squared: {r_squared:.4f})')

# Display legend outside the plot
plt.legend()

# Display plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define age column renaming mapping
column_name_mapping = {
    'pop_18_19': 'Age 18-19',
    'pop_20': 'Age 20',
    'pop_21': 'Age 21',
    'pop_22_24': 'Age 22-24',
    'pop_25_29': 'Age 25-29',
    'pop_30_34': 'Age 30-34',
    'pop_35_39': 'Age 35-39',
    'pop_40_44': 'Age 40-44',
    'pop_45_49': 'Age 45-49',
    'pop_50_54': 'Age 50-54',
    'pop_55_59': 'Age 55-59',
    'pop_60_61': 'Age 60-61',
    'pop_62_64': 'Age 62-64'
}

# Update the column names in merged_df
population_cols = list(column_name_mapping.keys())

# Group by zip code and calculate the mean (average) of chosen columns
grouped_df = merged_df.groupby('zcta')[population_cols].agg(np.mean).reset_index()

# Define custom colors for the bar chart
colors = ['#1E49E2', '#00B8F5', '#76D2FF', '#ACEAFF', '#510DBC', '#7213EA', '#B497FF',
          '#AB0D82', '#FD349C', '#FFA3DA', '#098E7E', '#00C0AE', '#63EBDA', '#333333']

# Create a bar chart with race percentage of the zip codes, using custom colors
ax = grouped_df[population_cols].plot(kind='bar', stacked=True, figsize=(12, 6), color=colors)

# Add labels and title
plt.xlabel('Zip Codes')
plt.ylabel('Average Population')
plt.title('Bronx Zip Code vs Average Age Population')

# Set x-tick labels to zip codes of the grouped data
ax.set_xticklabels(grouped_df['zcta'])

# Rotate the x-axis labels
plt.xticks(rotation=45, ha='right')

# Display legend with descriptive column names
legend_labels = [column_name_mapping[col] for col in population_cols]
plt.legend(legend_labels, bbox_to_anchor=(1.01, 1.0), loc='upper left')

# Display chart
plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Define age column renaming mapping
column_name_mapping = {
    'pop_18_19': 'Age 18-19',
    'pop_20': 'Age 20',
    'pop_21': 'Age 21',
    'pop_22_24': 'Age 22-24',
    'pop_25_29': 'Age 25-29',
    'pop_30_34': 'Age 30-34',
    'pop_35_39': 'Age 35-39',
    'pop_40_44': 'Age 40-44',
    'pop_45_49': 'Age 45-49',
    'pop_50_54': 'Age 50-54',
    'pop_55_59': 'Age 55-59',
    'pop_60_61': 'Age 60-61',
    'pop_62_64': 'Age 62-64'
}

# Update the column names in merged_df
population_cols = list(column_name_mapping.keys())

# Prepare input data (X) and target variables (y)
X = merged_df['zcta'].values.reshape(-1, 1)
y = merged_df[population_cols]

# Split the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Add a constant term to the input data for statsmodels
X_train_sm = sm.add_constant(X_train)
X_test_sm = sm.add_constant(X_test)

# Fit the data using Ordinary Least Squares (OLS) linear regression for each age column
results = []
for col in population_cols:
    model = sm.OLS(y_train[col], X_train_sm).fit()
    results.append(model)

# Print the summary of the regression models
for idx, result in enumerate(results):
    print("Model for age group:", column_name_mapping[population_cols[idx]])
    print(result.summary())
    print("\n")

# Alternatively, using scikit-learn for multivariate linear regression
sklearn_model = LinearRegression()
sklearn_model.fit(X_train, y_train)

# Make predictions using the test data
y_pred = sklearn_model.predict(X_test)

# Calculate and print Mean Squared Errors
mse_list = []
for idx, col in enumerate(population_cols):
    mse = mean_squared_error(y_test[col], y_pred[:, idx])
    mse_list.append(mse)
    print(f"Mean squared error for age group {column_name_mapping[col]}: {mse}")

# Calculate and print R-squared scores
r2_list = []
for idx, col in enumerate(population_cols):
    r2 = r2_score(y_test[col], y_pred[:, idx])
    r2_list.append(r2)
    print(f"R-squared score for age group {column_name_mapping[col]}: {r2}")
```
