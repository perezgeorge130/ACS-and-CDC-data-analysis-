```python
import pandas as pd

# Display data in the CDC file with zip code filter that removes zip codes less than 5 digits
df = pd.read_csv("CDC Places 2023 Health Outcomes.csv")
df = df[df['zcta5']>= 10000]
df
```

```python
# Filter CDC data for desired columns
filtered_columns = df[['zcta5', 'casthma_crudeprev', 'casthma_crude95ci', 'standardized_date']]
#dfrename = filtered_columns.rename(columns={'zcta5': 'Zip Code', 'casthma_crudeprev	': 'Asthma in adults of >=18 years of age(2021)', 'casthma_crude95ci' : 'Estimated confidence interval for asthma in adults of >=18 years of age (2021)'})
#dfrename
```

```python
# Read and filter ACS data file for desired county
df2 = pd.read_csv('ACS Socio-Demographics Curation_2023.csv')
filtered_df2 = df2[df2['county_name']=='Bronx County']
filtered_df2
```

```python
# Merge CDC and ACS data file based on zip code
merged_df = pd.merge(filtered_columns, filtered_df2, left_on='zcta5', right_on='zip_code', how='inner')
merged_df
```

```python
# Filter merged dataframe for desired columns (these will be used to create figures)
filtered_merged_df = merged_df[['zcta', 'casthma_crudeprev', 'casthma_crude95ci', 'standardized_date', 'po_name', 'county_name', 'poor_households_pct', 'hh_under_10k', 'hh_10k_15k',	'hh_15k_20k', 'hh_20k_25k',	'hh_25k_30k', 'hh_30k_35k',	'hh_35k_40k', 'hh_40k_45k',	'hh_45k_50k', 'hh_50k_60k']]
```

```python
# Read and filter phar data file for desired county
pharmacies_df = pd.read_csv('PharmacyInBronx.csv')
pharmacies_df
```

```python
# Merge CDC/ACS and pharmacies data file based on zip code
new_merged_df = pd.merge(filtered_merged_df, pharmacies_df, left_on='zcta', right_on='zip_code', how='inner')
new_merged_df
```

```python
# Create scatterplot for average asthma versus poor household percentage w/ line of best fit
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt

# Create a scatterplot
plt.scatter(new_merged_df['casthma_crudeprev'], new_merged_df['poor_households_pct'])

# Add label and title
plt.xlabel('Average Asthma')
plt.ylabel('Number of Low-Income Households')
plt.title(f'Scatter Plot for Average Asthma vs Low-Income Households (The Bronx)')

# Calculate the coefficients (slope and intercept) for the line of best fit
coefficients = np.polyfit(filtered_merged_df['casthma_crudeprev'], filtered_merged_df['poor_households_pct'], 1)

# Extract the slope from the coefficients
slope = coefficients[0]

# Create the line of best fit using the coefficients
best_fit_line = np.poly1d(coefficients)

# Calculate the R-squared value
slope, intercept, r_value, p_value, std_err = linregress(new_merged_df['casthma_crudeprev'], new_merged_df['poor_households_pct'])
r_squared = r_value**2
print(f'R-squared value: {r_squared:.4f}')

# Plot the line of best fit
plt.plot(new_merged_df['casthma_crudeprev'], best_fit_line(new_merged_df['casthma_crudeprev']), color='red', label=f'Line of Best Fit (Slope: {slope:.4f}), R-squared: {r_squared:.04})')

# Display legend
plt.legend()

# Display plot
plt.show()
```
