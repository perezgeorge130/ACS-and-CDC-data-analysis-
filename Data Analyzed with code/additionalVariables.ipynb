```python
import pandas as pd

# Display data in the CDC file with zip code filter that removes zip codes less than 5 digits
df = pd.read_csv("CDC Places 2023 Health Outcomes.csv")
df = df[df['zcta5']>= 10000]
df
```

```python
# Read and filter ACS data file for desired county
df2 = pd.read_csv('ACS Socio-Demographics Curation_2023.csv')
filtered_df2 = df2[df2['county_name'] == 'Bronx County']
filtered_df2
```

```python
# Merge CDC and ACS data file based on zip code
merged_df = pd.merge(df, filtered_df2, left_on='zcta5', right_on='zip_code', how='inner')
merged_df
```

```python
# Create scatterplot for average asthma versus poor household percentage w/ line of best fit
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt

# Create a scatterplot
plt.scatter(merged_df['hh_rented_pct'], merged_df['casthma_crudeprev'], color='#00338D')

# Add label and title
plt.xlabel('Average Renting')
plt.ylabel('Average Asthma')
plt.title(f'Scatter Plot for Average Asthma vs Average Renting')

# Calculate the coefficients (slope and intercept) for the line of best fit
coefficients = np.polyfit(merged_df['hh_rented_pct'], merged_df['casthma_crudeprev'], 1)

# Extract the slope from the coefficients
slope = coefficients[0]

# Create the line of best fit using the coefficients
best_fit_line = np.poly1d(coefficients)

# Calculate the R-squared value
slope, intercept, r_value, p_value, std_err = linregress(merged_df['hh_rented_pct'], merged_df['casthma_crudeprev'])
r_squared = r_value**2
print(f'R-squared value: {r_squared:.4f}')

# Plot the line of best fit
plt.plot(merged_df['hh_rented_pct'], best_fit_line(merged_df['hh_rented_pct']), color='#6D2077', label=f'Line of Best Fit (Slope: {slope:.4f}), R-squared: {r_squared:.04})')

# Display legend
plt.legend()

# Display plot
plt.show()
```

```python
import numpy as np
from scipy.stats import linregress
import pandas as pd
import matplotlib.pyplot as plt

# Define the lowest and highest income range columns
lowest_income_col = 'labor_force'
highest_income_col = 'not_in_labor_force'

# Extract the necessary data for the scatter plot
lowest_income_df = merged_df[merged_df[lowest_income_col].notnull()]
lowest_income_x = lowest_income_df[lowest_income_col]
lowest_income_y = lowest_income_df['casthma_crudeprev']

highest_income_df = merged_df[merged_df[highest_income_col].notnull()]
highest_income_x = highest_income_df[highest_income_col]
highest_income_y = highest_income_df['casthma_crudeprev']

# Create a scatter plot with two colors
plt.scatter(lowest_income_x, lowest_income_y, color='#00338D')
plt.scatter(highest_income_x, highest_income_y, color='#6D2077')

# Add labels and title to the plot
plt.xlabel(f'Population ({lowest_income_col} and {highest_income_col})')
plt.ylabel('Average Asthma')
plt.title(f'Scatter Plot for {lowest_income_col} vs {highest_income_col} with Average Asthma')

# Add R-squared lines for each income range
slope, intercept, r_value, p_value, std_err = linregress(lowest_income_x, lowest_income_y)
r_squared = r_value**2
print(f'Low Income Range R-squared: {r_squared}')
plt.plot(lowest_income_x, slope * lowest_income_x + intercept, color='#00338D', label=f'{lowest_income_col} (R-squared: {r_squared:.4f})')

slope, intercept, r_value, p_value, std_err = linregress(highest_income_x, highest_income_y)
r_squared = r_value**2
print(f'High Income Range R-squared: {r_squared}')
plt.plot(highest_income_x, slope * highest_income_x + intercept, color='#6D2077', label=f'{highest_income_col} (R-squared: {r_squared:.4f})')

# Display legend outside the plot
plt.legend()

# Display plot
plt.show()
```
