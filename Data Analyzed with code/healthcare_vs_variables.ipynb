```python
import pandas as pd

# Display data in the CDC file with zip code filter that removes zip codes less than 5 digits
df = pd.read_csv("CDC Places 2023 Health Outcomes.csv")
df = df[df['zcta5']>= 10000]
df
```

```python
# Read and filter ACS data file for desired county
df2 = pd.read_csv('ACS Socio-Demographics Curation_2023.csv')
filtered_df2 = df2[df2['county_name'] == 'Bronx County']
filtered_df2
```

```python
# Merge CDC and ACS data file based on zip code
merged_df = pd.merge(df, filtered_df2, left_on='zcta5', right_on='zip_code', how='inner')
merged_df
```

```python
# Create scatterplot for estimated access to healthcare versus average asthma w/ line of best fit
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt

# Create a scatterplot
plt.scatter(merged_df['disability_crudeprev'], merged_df['access2_crudeprev'])

# Add label and title
plt.xlabel('Average of disability')
plt.ylabel('Average Lack of Access to Healthcare')
plt.title(f'Scatter Plot for Average Lack of Access to Healthcare vs Average of  disability')

# Calculate the coefficients (slope and intercept) for the line of best fit
coefficients = np.polyfit(merged_df['disability_crudeprev'], merged_df['access2_crudeprev'], 1)

# Extract the slope from the coefficients
slope = coefficients[0]

# Create the line of best fit using the coefficients
best_fit_line = np.poly1d(coefficients)

# Calculate the R-squared value
slope, intercept, r_value, p_value, std_err = linregress(merged_df['disability_crudeprev'], merged_df['access2_crudeprev'])
r_squared = r_value**2
print(f'R-squared value: {r_squared:.4f}')

# Plot the line of best fit
plt.plot(merged_df['disability_crudeprev'], best_fit_line(merged_df['disability_crudeprev']), color='red', label=f'Line of Best Fit (Slope: {slope:.4f}), R-squared: {r_squared:.04})')

# Display legend
plt.legend()

# Display plot
plt.show()
```
