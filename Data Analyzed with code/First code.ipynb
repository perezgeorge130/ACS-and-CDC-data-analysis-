```python
import pandas as pd

# Display data in the CDC file with zip code filter that removes zip codes less than 5 digits
df = pd.read_csv("CDC Places 2023 Health Outcomes.csv")
df = df[df['zcta5']>= 10000]
df
```

```python
# Read and filter ACS data file for desired county
df2 = pd.read_csv('ACS Socio-Demographics Curation_2023.csv')
filtered_df2 = df2[df2['county_name'] == 'Bronx County']
filtered_df2
```

```python
# Merge CDC and ACS data file based on zip code
merged_df = pd.merge(df, filtered_df2, left_on='zcta5', right_on='zip_code', how='inner')
merged_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create bar chart using desired columns
plt.bar(merged_df['zcta'], merged_df['casthma_crudeprev'])

# Add labels and title
plt.xlabel('Zip Codes')
plt.ylabel('Average Asthma Prevalence')
plt.title(f'Bronx Zip Code vs Average Asthma Prevalence')

# Set the axis limits
plt.ylim(0, 18)

# Plot a horzontal line for the average
average_value = merged_df['casthma_crudeprev'].mean()
plt.axhline(y=average_value, color = 'red', linestyle = '--', label = f'Average ({average_value:.2f})')

# Rotate the x-axis labels
plt.xticks(rotation=45, ha='right')

# Display legend
plt.legend()

# Display chart
plt.show()
```

```python
# Create scatterplot for average asthma versus poor household percentage w/ line of best fit
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt

# Create a scatterplot
plt.scatter(merged_df['poor_households_pct'], merged_df['casthma_crudeprev'] , color='#00338D')

# Add label and title
plt.xlabel('Poor Household Percentage')
plt.ylabel('Average Asthma Prevalence')
plt.title(f'Average Asthma Prevalence and Poor Households')

# Calculate the coefficients (slope and intercept) for the line of best fit
coefficients = np.polyfit(merged_df['poor_households_pct'], merged_df['casthma_crudeprev'], 1)

# Extract the slope from the coefficients
slope = coefficients[0]

# Create the line of best fit using the coefficients
best_fit_line = np.poly1d(coefficients)

# Calculate the R-squared value
slope, intercept, r_value, p_value, std_err = linregress(merged_df['poor_households_pct'], merged_df['casthma_crudeprev'])
r_squared = r_value**2
print(f'R-squared value: {r_squared:.4f}')

# Plot the line of best fit #005EB8
plt.plot(merged_df['poor_households_pct'], best_fit_line(merged_df['poor_households_pct']), color='#0091DA', label=f'Line of Best Fit (Slope: {slope:.4f}), R-squared: {r_squared:.04})')

# Display legend
plt.legend()

# Display plot
plt.show()
```
