```python
import pandas as pd

# Display data in the CDC file with zip code filter that removes zip codes less than 5 digits
df = pd.read_csv("CDC Places 2023 Health Outcomes.csv")
df = df[df['zcta5']>= 10000]
df
```

```python
# Read and filter ACS data file for desired county
df2 = pd.read_csv('ACS Socio-Demographics Curation_2023.csv')
merged_df2 = df2[df2['county_name'] == 'Bronx County']
merged_df2
```

```python
# Merge CDC and ACS data file based on zip code
merged_df = pd.merge(df, merged_df2, left_on='zcta5', right_on='zip_code', how='inner')
merged_df
```

```python
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Define the column mapping for transportation columns
transportation_col_mapping = {
    #Group 1
    'transp_public_to_work': 'Public Transportation to Work',
    'transp_car_to_work': 'Car to Work',

    #Group 2
    'transp_bicycle_to_work': 'Bicycle to Work',
    'transp_motorcycle_to_work': 'Motorcycle to Work',
    'transp_other_to_work': 'Other to Work',
    'transp_taxi_to_work': 'Taxi to Work',

    #Group 3
    'transp_walked_to_work': 'Walked to Work'
}

# Define transportation columns to use
transportation_cols = list(transportation_col_mapping.keys())

# Find the maximum number of columns within the groups
n_columns = max([2, 3, 1])

# Create a figure and define the layout of the subplots (2 rows, 4 columns)
fig, axs = plt.subplots(2, 4, figsize=(24, 10), sharey=True)
axs = axs.flat

# Fill the remaining subplots with empty plots (this keeps the grid structure)
for i in range(len(transportation_cols), len(axs)):
    axs[i].axis('off')

# Create the scatterplot for each transportation column
for i, col in enumerate(transportation_cols):
    # Create a scatterplot
    axs[i].scatter(merged_df[col], merged_df['casthma_crudeprev'], label=col, color='#00338D')

    # Add a red regression line
    slope, intercept = np.polyfit(merged_df[col], merged_df['casthma_crudeprev'], 1)
    regression_line = np.poly1d([slope, intercept])
    line = Line2D([], [], color='#6D2077', label='R-squared')
    axs[i].plot(merged_df[col], regression_line(merged_df[col]), color='#6D2077')
    
    # Add label and title
    axs[i].set_title(f'Average Asthma Prevalence vs {transportation_col_mapping[col]}')
    axs[i].set_xlabel(f'{transportation_col_mapping[col]}')
    axs[i].set_ylabel('Average Asthma Prevalence')

    # Calculate the R-squared value
    slope, intercept, r_value, p_value, std_err = linregress(merged_df[col], merged_df['casthma_crudeprev'])
    r_squared = r_value**2
    print(f'R-squared value for {transportation_col_mapping[col]}: {r_squared:.4f}')

    # Add the R-squared value to the legend
    axs[i].legend(handles=[line], labels=[f'R-squared: {r_squared:.4f}'], loc='upper left')

# Adjust the layout
plt.tight_layout()

# Display plot
plt.show()
```

```python
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Define the column mapping for transportation columns
transportation_col_mapping = {
    # Group 1
    'transp_public_to_work': 'Public Transportation',
    'transp_car_to_work': 'Car',

    # Group 2
    'transp_bicycle_to_work': 'Bicycle',
    'transp_motorcycle_to_work': 'Motorcycle',
    'transp_other_to_work': 'Other',
    'transp_taxi_to_work': 'Taxi',

    # Group 3
    'transp_walked_to_work': 'Walked'
}

# Define transportation column groups
vehicle_group = ['transp_public_to_work', 'transp_car_to_work', 'transp_bicycle_to_work', 'transp_motorcycle_to_work', 'transp_taxi_to_work']
other_walking_group = ['transp_other_to_work', 'transp_walked_to_work']

# Group names and colors
groups = [(vehicle_group, "Vehicles"), (other_walking_group, "Other/Walking")]

# Create a scatterplot for each group
fig, axs = plt.subplots(1, 2, figsize=(15, 6), sharey=True)
grouped_df = merged_df.copy()
grouped_df['Vehicle'] = merged_df[vehicle_group].sum(axis=1)
grouped_df['Other_Walking'] = merged_df[other_walking_group].sum(axis=1)

for ax, (group, group_name) in zip(axs, groups):
    # Create a scatterplot using the total value of each group
    if group_name == "Vehicles":
        ax.scatter(grouped_df['Vehicle'], grouped_df['casthma_crudeprev'], label=group_name, color='#00338D')
        x = grouped_df['Vehicle']
    else:
        ax.scatter(grouped_df['Other_Walking'], grouped_df['casthma_crudeprev'], label=group_name, color='#00338D')
        x = grouped_df['Other_Walking']
      
    y = grouped_df['casthma_crudeprev']

    # Add a regression line
    slope, intercept = np.polyfit(x, y, 1)
    regression_line = np.poly1d([slope, intercept])
    ax.plot(x, regression_line(x), color='#0091DA', linestyle='--')

    # Calculate the R-squared value
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    r_squared = r_value**2
    print(f'R-squared value for {group_name}: {r_squared:.4f}')

    ax.set_title(f'{group_name}: Asthma Prevalence vs. Transportation Methods')
    ax.set_ylabel('Average Asthma Prevalence')
    ax.set_xlabel('Count of People')
    ax.legend()

# Adjust the layout
plt.tight_layout()

# Display plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Define the column mapping for transportation columns
transportation_col_mapping = {
    #Group 1
    'transp_public_to_work': 'Public Transportation',
    'transp_car_to_work': 'Car',

    #Group 2
    'transp_bicycle_to_work': 'Bicycle',
    'transp_motorcycle_to_work': 'Motorcycle',
    'transp_other_to_work': 'Other',
    'transp_taxi_to_work': 'Taxi',

    #Group 3
    'transp_walked_to_work': 'Walked'
}

# Define transportation columns to use
transportation_cols = list(transportation_col_mapping.keys())

# Prepare input data (X) and target variables (y)
X = merged_df['zcta'].values.reshape(-1, 1)
y = merged_df[transportation_cols]

# Split the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Add a constant term to the input data for statsmodels
X_train_sm = sm.add_constant(X_train)
X_test_sm = sm.add_constant(X_test)

# Fit the data using Ordinary Least Squares (OLS) linear regression for each transportation column
results = []
for col in transportation_cols:
    model = sm.OLS(y_train[col], X_train_sm).fit()
    results.append(model)

# Print the summary of the regression models
for idx, result in enumerate(results):
    print("Model for income group:", transportation_col_mapping[transportation_cols[idx]])
    print(result.summary())
    print("\n")

# Alternatively, using scikit-learn for multivariate linear regression
sklearn_model = LinearRegression()
sklearn_model.fit(X_train, y_train)

# Make predictions using the test data
y_pred = sklearn_model.predict(X_test)

# Calculate and print Mean Squared Errors
mse_list = []
for idx, col in enumerate(transportation_cols):
    mse = mean_squared_error(y_test[col], y_pred[:, idx])
    mse_list.append(mse)
    print(f"Mean squared error for income group {transportation_col_mapping[col]}: {mse}")

# Calculate and print R-squared scores
r2_list = []
for idx, col in enumerate(transportation_cols):
    r2 = r2_score(y_test[col], y_pred[:, idx])
    r2_list.append(r2)
    print(f"R-squared score for income group {transportation_col_mapping[col]}: {r2}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define the column mapping for transportation columns
transportation_col_mapping = {
    # Group 1
    'transp_public_to_work': 'Public Transportation',
    'transp_car_to_work': 'Car',

    # Group 2
    'transp_bicycle_to_work': 'Bicycle',
    'transp_motorcycle_to_work': 'Motorcycle',
    'transp_other_to_work': 'Other',
    'transp_taxi_to_work': 'Taxi',

    # Group 3
    'transp_walked_to_work': 'Walked'
}

# Define transportation column groups
group1 = ['transp_public_to_work', 'transp_car_to_work', 'transp_walked_to_work']
group2 = ['transp_bicycle_to_work', 'transp_motorcycle_to_work', 'transp_other_to_work', 'transp_taxi_to_work']

# Group by zip code and calculate the mean of chosen columns and asthma prevalence
grouped_df = merged_df.groupby('zcta')[transportation_cols + ['casthma_crudeprev']].agg(np.mean).reset_index()

# Function to create a grouped bar chart for a given group of columns
def create_grouped_bar_chart(group, title, ylabel='Average Asthma Prevalence', xlabel_rotation=0):
    # Define the number of bars and the width of each bar
    n_bars = len(group)
    bar_width = 0.8 / n_bars

    # Define custom colors for the bar chart
    colors = ['#00338D', '#ACEAFF', '#B497FF', '#FD349C']

    # Create a bar chart for each column in the group with spacing between bars
    for i, col in enumerate(group):
        plt.bar(grouped_df.index + (i * bar_width), grouped_df[col], width=bar_width, color=colors[i % len(colors)], label=transportation_col_mapping[col])

    # Add labels and title
    plt.xlabel('Zip Codes')
    plt.ylabel(ylabel)
    plt.title(title)

    # Set x-tick labels to zip codes of the grouped data
    plt.xticks(grouped_df.index + (bar_width * (n_bars - 1) / 2), grouped_df['zcta'], rotation=xlabel_rotation, ha='right')

    # Display legend
    plt.legend(bbox_to_anchor=(1.01, 1.0), loc='upper left')

# Group 1 graph
plt.figure(figsize=(12, 6))
create_grouped_bar_chart(group1, 'Group 1: Public Transportation, Car, and Walked', 'Population Count', xlabel_rotation=45)
plt.show()

# Group 2 graph
plt.figure(figsize=(12, 6))
create_grouped_bar_chart(group2, 'Group 2: Taxi, Other, Motorcycle, Bicycle', 'Population Count', xlabel_rotation=45)
plt.show()
```
