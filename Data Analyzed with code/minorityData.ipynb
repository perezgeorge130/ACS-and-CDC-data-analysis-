```python
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt

# Create a scatterplot with the specified color
plt.scatter(merged_df['casthma_crudeprev'], merged_df['am_indian_pct'], color='#1E49E2')

# Add label and title
plt.xlabel('Average Asthma Prevalence')
plt.ylabel('Percentage of American Indian People')
plt.title(f'Average Asthma Prevalence vs American Indian Population')

# Calculate the coefficients (slope and intercept) for the line of best fit
coefficients = np.polyfit(merged_df['casthma_crudeprev'], merged_df['am_indian_pct'], 1)

# Extract the slope from the coefficients
slope = coefficients[0]

# Create the line of best fit using the coefficients
best_fit_line = np.poly1d(coefficients)

# Calculate the R-squared value
slope, intercept, r_value, p_value, std_err = linregress(merged_df['casthma_crudeprev'], merged_df['am_indian_pct'])
r_squared = r_value**2
print(f'R-squared value: {r_squared:.4f}')

# Plot the line of best fit with the specified color
plt.plot(merged_df['casthma_crudeprev'], best_fit_line(merged_df['casthma_crudeprev']), color='#0C233C', label=f'Line of Best Fit (Slope: {slope:.4f}, R-squared: {r_squared:.4f})')

# Display legend
plt.legend()

# Display plot
plt.show()
```

```python
import numpy as np
from scipy.stats import linregress
import pandas as pd
import matplotlib.pyplot as plt

# Define column renaming mapping
column_name_mapping = {
    'am_indian_pct': 'American Indian Percentage',
    'asian_pct': 'Asian Percentage',
    'black_pct': 'Black Percentage',
    'hawaiian_pct': 'Hawaiian Percentage',
    'white_pct': 'White Percentage',
    'minority_pct': 'Minority Percentage'
}

income_cols = list(column_name_mapping.keys())

# Define the number of rows and columns for the plot matrix
n_rows = 2  # Change to the desired number of rows
n_cols = 3  # Change to the desired number of columns

# Create the figure and the plot matrix
fig, axs = plt.subplots(n_rows, n_cols, figsize=(22, 10))

# Define the colors for the scatter points and the line of best fit
scatter_color = '#1E49E2'
line_color = '#470A68'

# Iterate through each cell of the plot matrix
for i in range(n_rows):
    for j in range(n_cols):
        # Calculate the index of the income column to use
        k = i * n_cols + j

        # Check if the index is valid, i.e., within the range of income_cols and the column has non-null values
        if k < len(income_cols) and not merged_df[income_cols[k]].isnull().all():
            # Extract the data for the scatter plot
            x = merged_df[income_cols[k]]
            y = merged_df['casthma_crudeprev']

            # Create a scatter plot with specified color
            axs[i, j].scatter(x, y, c=scatter_color)

            # Add label and title for each plot
            axs[i, j].set_xlabel('Race Percentage')
            axs[i, j].set_ylabel('Prevalence of Asthma')
            axs[i, j].set_title(f'Average Asthma Prevalence for {column_name_mapping[income_cols[k]]}')

            # Calculate the coefficients (slope and intercept) for the line of best fit
            coefficients = np.polyfit(x, y, 1)

            # Extract the slope from the coefficients
            slope = coefficients[0]

            # Calculate the R-squared value
            slope, intercept, r_value, p_value, std_err = linregress(x, y)
            r_squared = r_value ** 2
            print(f'{column_name_mapping[income_cols[k]]} - R-Squared: {r_squared}')

            # Add the line of best fit to the scatter plot with specified color
            axs[i, j].plot(x, slope * x + intercept, color=line_color, label=f'R-squared: {r_squared:.4f}')

            # Display legend
            axs[i, j].legend()

            # Add a light grey grid to each subplot
            # axs[i, j].grid(True, color='lightgray', alpha=0.7)

# Adjust the padding between subplots
fig.tight_layout(pad=2)

# Display plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Define column renaming mapping
column_name_mapping = {
    'am_indian_pct': 'American Indian',
    'asian_pct': 'Asian',
    'black_pct': 'Black',
    'hawaiian_pct': 'Hawaiian',
    'white_pct': 'White',
}

# Define income columns to use without 'minority_pct'
income_cols = list(column_name_mapping.keys())

# Prepare input data (X) and target variables (y)
X = merged_df['zcta'].values.reshape(-1, 1)
y = merged_df[income_cols]

# Split the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Add a constant term to the input data for statsmodels
X_train_sm = sm.add_constant(X_train)
X_test_sm = sm.add_constant(X_test)

# Fit the data using Ordinary Least Squares (OLS) linear regression for each income column
results = []
for col in income_cols:
    model = sm.OLS(y_train[col], X_train_sm).fit()
    results.append(model)

# Print the summary of the regression models
for idx, result in enumerate(results):
    print("Model for income group:", column_name_mapping[income_cols[idx]])
    print(result.summary())
    print("\n")

# Alternatively, using scikit-learn for multivariate linear regression
sklearn_model = LinearRegression()
sklearn_model.fit(X_train, y_train)

# Make predictions using the test data
y_pred = sklearn_model.predict(X_test)

# Calculate and print Mean Squared Errors
mse_list = []
for idx, col in enumerate(income_cols):
    mse = mean_squared_error(y_test[col], y_pred[:, idx])
    mse_list.append(mse)
    print(f"Mean squared error for income group {column_name_mapping[col]}: {mse}")

# Calculate and print R-squared scores
r2_list = []
for idx, col in enumerate(income_cols):
    r2 = r2_score(y_test[col], y_pred[:, idx])
    r2_list.append(r2)
    print(f"R-squared score for income group {column_name_mapping[col]}: {r2}")
```
